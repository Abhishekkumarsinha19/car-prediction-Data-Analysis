```python
def convert_to_float(value):
    try:
        if isinstance(value, float) or isinstance(value, int):
            return float(value)
        if pd.isna(value) or value.strip() == '':
            return None
        clean_value = ''.join(filter(lambda x: x.isdigit() or x == '.', str(value)))
        return float(clean_value) if clean_value else None
    except ValueError:
        return None

data['Mileage'] = data['Mileage'].apply(convert_to_float)
data['Power'] = data['Power'].apply(convert_to_float)
data.head()
```

```python
def convert_to_int(value):
    try:
        if isinstance(value, float) or isinstance(value, int):
            return int(value)
        if pd.isna(value) or value.strip() == '':
            return None
        clean_value = ''.join(filter(lambda x: x.isdigit(), str(value)))
        return int(clean_value) if clean_value else None
    except ValueError:
        return None
data['Engine'] = data['Engine'].apply(convert_to_int)
data.head()
```

```python
# Function for log transformation of the column - Log transformation can help in normalization to maintain the standard scaling with other variables
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

```python
log_transform(data,['Kilometers_Driven','Price'])
```
